```python
import pandas as pd
import numpy as np

df = pd.read_csv('../../datasets/raw_malt_data.csv')
df
```

```python
# separate the data into two dataframes

# if the name is NaN, then the data is not scrapped
df_data_scrapped = df[df['name'].notna()]
df_profile_links_not_scrapped = df[df['name'].isna()]

df_data_scrapped.to_csv('../../datasets/malt_data_scrapped.csv', index=False)
df_data_scrapped
```

```python
# keep only link, creation_date and name

df_profile_links_not_scrapped = df_profile_links_not_scrapped[['creation_date', 'profil', 'link']]
df_profile_links_not_scrapped.to_csv('../../datasets/malt_profile_links_not_scrapped.csv', index=False)

df_profile_links_not_scrapped
```
