```python
import pandas as pd

df_parsed = pd.read_csv('../../datasets/parsed_malt_data.csv')
# df_parsed = df_parsed[:200]
```

```python
import ast
df_cities = pd.read_csv('../../datasets/cities_list.csv')

# Convert the string 'location' column to a dictionary and extract the city
def extract_city(location_str):
    try:
        location_dict = ast.literal_eval(location_str)
        city_country = location_dict.get('Localisation', '')
        city = city_country.split(',')[0].strip()
        return city
    except:
        return None

df_parsed['city'] = df_parsed['location'].apply(extract_city)
df_parsed
```

NEW FIND REGION NAME

```python
def get_region_name(city, df_cities):
    if city is None:
        return None
    df_city = df_cities[df_cities['label'].str.lower() == city.lower()]
    if len(df_city) == 0:
        return None
    return df_city.iloc[0]['region_name']

df_parsed['Region_Name'] = df_parsed.apply(lambda row: get_region_name(row['city'], df_cities), axis=1)

# Remove lines where 'Region_Name' is missing
df_parsed.dropna(subset=['Region_Name'], inplace=True)
```

```python
# add timestamp column
df_parsed['timestamp'] = pd.to_datetime(df_parsed['creation_date'])
```

```python
df_parsed.to_csv('../../datasets/final_parsed_malt_data.csv', index=False)

df_parsed
```
