```python
# import modules
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np

df = pd.read_csv('../../datasets/parsed_malt_data.csv')
df
```

## USER COUNT BY YEAR

```python
df['creation_date'] = pd.to_datetime(df['creation_date'], format='%Y-%m-%d')

# get only the year from the date
vis = df['creation_date'].dt.year.value_counts() 

# create a new figure
vis.plot()

# sort the data by index
vis.sort_index(ascending=True, inplace=True)

# add a title and labels
plt.title('User Count by Year')
plt.xlabel('Year')
plt.ylabel('User Count')

# show the plot
plt.show()
```

```python
# Supposons que votre fichier CSV a des colonnes 'creation_date'
# Convertir la colonne 'creation_date' en format de date
df['creation_date'] = pd.to_datetime(df['creation_date'], format='%Y-%m-%d')

# Sélectionner les 4 premiers caractères de chaque élément dans la colonne 'creation_date'
df['annee'] = df['creation_date'].dt.year.astype(str).str[:4]

# Compter le nombre d'occurrences pour chaque année
occurrences_par_annee = df['annee'].value_counts()

# Trier le DataFrame par année
occurrences_par_annee = occurrences_par_annee.sort_index()

# Créer un diagramme à barres
occurrences_par_annee.plot(kind='bar', color='skyblue')

# Ajouter des titres et des labels
plt.title('Nombre d\'occurrences par année')
plt.xlabel('Année')
plt.ylabel('Nombre d\'occurrences')

# Afficher le diagramme
plt.show()
```
