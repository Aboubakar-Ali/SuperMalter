```python
# import modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import numpy as np

df = pd.read_csv('./../../../final_parsed_malt_data_cleaned.csv')
df.head()
```

```python
print(df.columns)
```

```python
# Recommandations per profile
recommandations_data = df["recommendations"]

# Select the top 5 regions
top_recommandations = recommandations_data.head(5)

plt.figure(figsize=(10, 6))
top_recommandations.plot(kind='bar', color='coral')
plt.title('Recommandations per profile (head 5)')
plt.xlabel('profil')
plt.ylabel('Recommandations')
plt.xticks(rotation=45)
plt.show()
```

```python
# Affichage des 5 premières recommandations par profil sous forme de graphique à barres
plt.figure(figsize=(10, 6))
recommandations_data.plot(kind='bar', color='coral')
plt.title('Recommandations per profile (Top 5)')
plt.xlabel('Profile')
plt.ylabel('Recommandations')
plt.xticks(rotation=45)
plt.show()


filtered_df = df[df['recommendations'] != 0]
recommandations_mean = filtered_df['recommendations'].mean()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.reset_index(), x=df.index, y='recommendations', color='blue', s=50)
plt.axhline(y=recommandations_mean, color='red', linestyle='--', label=f'Moyenne: {recommandations_mean:.2f}')
plt.title('Recommandations par profil et moyenne')
plt.xlabel('Profil')
plt.ylabel('Recommandations')
plt.ylim(0, 80)
plt.legend()
plt.show()

# sns.lmplot(
#     data=df, x='profile', y='recommendations',
#     palette="muted", ci=None,
#     height=6, scatter_kws={"s": 50, "alpha": 1}
# )
# plt.title('Régression linéaire entre Profile et Recommendations')
# plt.xlabel('Profile')
# plt.ylabel('Recommendations')
# plt.show()

```
